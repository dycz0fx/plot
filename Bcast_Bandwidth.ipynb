```python
%matplotlib inline
import matplotlib.pyplot as plt
import csv

x = []
ompi_topo_adapt = []
ompi_topo_tuned = []
ompi_default_tuned = []
mvapich2 = []
intel = []
p2p_64K = []
p2p = []

with open('Bcast_Bandwidth.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        print(row)
        if plots.line_num > 1:
            x.append(int(row[0]))
            ompi_topo_adapt.append(float(row[1]))
            ompi_topo_tuned.append(float(row[2]))
            ompi_default_tuned.append(float(row[3]))
            mvapich2.append(float(row[4]))
            intel.append(float(row[5]))
            p2p_64K.append(float(row[6]))
            p2p.append(float(row[7]))
            
fig = plt.figure(figsize=(15,10))
fig.suptitle('Bcast np128', fontsize=20)
ax1 = fig.add_subplot(111)
ax1.plot(x, ompi_topo_adapt, label='OMPI(topo+adapt)')
ax1.plot(x, ompi_topo_tuned, label='OMPI(topo+tuned)')    
ax1.plot(x, ompi_default_tuned, label='OMPI(default tuned)')
ax1.plot(x, mvapich2, label='MVAPICH2')
ax1.plot(x, intel, label='IntelMPI')
ax1.plot(x, p2p_64K, label='P2P(64k)')
ax1.plot(x, p2p, label='P2P')
ax1.set_ybound(lower=0,upper=12)
ax1.set_xlabel('Message Size (Bytes)', fontsize=15)
ax1.set_ylabel('Bandwidth (Gbps)', fontsize=15)
ax1.legend(loc=0,fontsize=12)

ax2 = ax1.twinx()
ax2.set_ylabel('Percentage', fontsize=15)


```
