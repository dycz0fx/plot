```python
%matplotlib inline
import matplotlib.pyplot as plt
import csv

x = []
tuned = []
adapt_2K = []
adapt_4K = []
adapt_8K = []
adapt_16K = []
adapt_32K = []
adapt_64K = []
adapt_128K = []
adapt_256K = []
adapt_512K = []
adapt_curvefit = []
adapt_loop = []
table = []

with open('bcast_chain_diffseg.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        print(row)
        table.append(row);
        if plots.line_num > 1:
            x.append(int(row[0]))
            tuned.append(float(row[1]))
            adapt_2K.append(float(row[2]))
            adapt_4K.append(float(row[3]))
            adapt_8K.append(float(row[4]))
            adapt_16K.append(float(row[5]))
            adapt_32K.append(float(row[6]))
            adapt_64K.append(float(row[7]))
            adapt_128K.append(float(row[8]))
            adapt_256K.append(float(row[9]))
            adapt_512K.append(float(row[10]))
            adapt_curvefit.append(float(row[11]))
            adapt_loop.append(float(row[12]))
```

```python
plt.figure(figsize=(15,10))
plt.title("Bcast Chain np=8, message 64K to 256K", fontsize=15)
plt.plot(x[0:3], adapt_8K[0:3], label = "adapt_8K")
plt.plot(x[0:3], adapt_64K[0:3], label = "adapt_64K")
plt.plot(x[0:3], adapt_512K[0:3], label = "adapt_256K")
plt.plot(x[0:3], adapt_curvefit[0:3], label = "adapt_curvefit")
plt.plot(x[0:3], adapt_loop[0:3], label = "adapt_loop")
plt.legend(loc=0,fontsize=12)
plt.xlabel('Message Size (bytes)', fontsize=15)
plt.ylabel('Latency (us)', fontsize=15)
```

```python
plt.figure(figsize=(15,10))
plt.title("Bcast Chain np=8 message 512K to 4M", fontsize=15)
plt.plot(x[3:], adapt_8K[3:], label = "adapt_8K")
plt.plot(x[3:], adapt_64K[3:], label = "adapt_64K")
plt.plot(x[3:], adapt_512K[3:], label = "adapt_512K")
plt.plot(x[3:], adapt_curvefit[3:], label = "adapt_curvefit")
plt.plot(x[3:], adapt_loop[3:], label = "adapt_loop")
plt.legend(loc=0,fontsize=12)
plt.xlabel('Message Size (bytes)', fontsize=15)
plt.ylabel('Latency (us)', fontsize=15)
```
